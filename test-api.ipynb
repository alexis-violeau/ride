```python
%load_ext autoreload
%autoreload 2

import requests
from api import BASE_URL, STATION_INFO
import geopandas as gpd
import pandas as pd
from sql_utils import get_db_connection
```

# Connect to DB

```python
con = get_db_connection()
```

# Load data from API

```python
answer = requests.get(BASE_URL + STATION_INFO)
```

# Format data

```python
stations = pd.json_normalize(answer.json(), record_path=['data', 'stations'])
```

```python
gpd_stations = gpd.GeoDataFrame(data = stations, geometry = gpd.points_from_xy(stations.lon, stations.lat), crs = "epsg:4326")
```

# Export in DB

```python
gpd_stations = gpd_stations.drop(columns=["lat","lon", "station_opening_hours", "rental_methods"])
```

```python
gpd_stations.to_postgis(con = con, name = "velib_stations")
```
